```python
data_dict = {'acc' : pd.DataFrame(),
             'bvp' : pd.DataFrame(),
             'eda' : pd.DataFrame(),
             'hr' : pd.DataFrame(),
             'temp' : pd.DataFrame()}
for file, location in {'ACC' : 'acc', 'BVP': 'bvp', 'EDA' : 'eda', 'HR' : 'hr', 'TEMP' : 'temp'}.items():
    for user in ['S13', 'S15', 'S04', 'S14', 'S07', 'S02', 'S03', 'S08', 'S09', 'S16', 'f09', 'f10', 'f05', 'f06', 'f07', 'S17', 'f08', 'user2', 'user3', 'user4', 'user5', 'user6', 'user8', 'user13', 'user7', 'user10', 'user12', 'user14', 'user16'] :
        data = pd.read_csv(f"C:\\Users\\roger\\Desktop\\New Complete\\{file}_{user}.csv", header = None)
        data = data.iloc[2:, :]
        if file == 'ACC':
            print(data.columns)
            data['movement'] = np.sqrt(data[0]**2 + data[1]**2 + data[2]**2)
            data['roll__mean_movement'] = data['movement'].rolling(window = 3).mean()
            data['roll_std_movement'] = data['movement'].rolling(window = 3).std()
            data['roll_skew_movement'] = data['movement'].rolling(window = 3).skew()
            data['roll_kurtosis_movement'] = data['movement'].rolling(window = 3).apply(lambda x : kurtosis(x, bias = False), raw = True)
            
        data_dict[location] = pd.concat([data_dict[location], data], axis = 0, ignore_index = True)

```

```python

data_dict['acc'].rename(columns = {'1' : 'acc_x', '2' : 'acc_y', '3' : 'acc_z'}, inplace = True)
data_dict['bvp'].rename(columns = {'0' : 'bvp'}, inplace = True)
data_dict['eda'].rename(columns = {'0' : 'eda'}, inplace = True)
data_dict['hr'].rename(columns = {'0' : 'hr'}, inplace = True)
data_dict['temp'].rename(columns = {'0' : 'temp'}, inplace = True)
```

```python
acc = data_dict['acc'].copy()
acc.dropna(inplace = True)
acc = acc.reset_index(drop = True)
bvp = data_dict['bvp'].copy()
bvp.dropna(inplace = True)
bvp = bvp.reset_index(drop = True)
eda = data_dict['eda'].copy()
eda.dropna(inplace = True)
eda = eda.reset_index(drop = True)
hr = data_dict['hr'].copy()
hr.dropna(inplace = True)
hr = hr.reset_index(drop = True)
temp = data_dict['temp'].copy()
temp.dropna(inplace = True)
temp = temp.reset_index(drop = True)
```

```python
actual_age = pd.concat([acc, bvp, eda, hr, temp], axis = 1, ignore_index = True)
print(np.sum(actual_age.isna()))
```
